```python
import pandas as pd
import numpy as np
```

```python
# Read in data
df = pd.read_csv("real_estate_data.csv")
```

```python
# Original metrics to gauge what to clean
print("Original shape\n: ", df.shape)
print("\nColumns: \n", df.columns)
print("\nData types: \n", df.dtypes)
print("\nNulls per column: \n", df.isnull().sum())
print("\nDuplicate count: \n", df.duplicated().sum())
```

```python
# Over 800,000 rows miss location or address, which we need, so they need to be dropped
df = df.dropna(subset=['Location', 'Address'])
print("Shape after cleaning missing locations and addresses: ", df.shape)
```

```python
# Non-usable codes "typically mean the sale price is not reliable for use in the determination of a property value", so drop that column and any rows that had a non-usable code

df = df[df['Non Use Code'].isna()]
df = df.drop(columns=['Non Use Code'])
print("Shape after dropping rows with non-usable codes: ", df.shape)
```

```python
# Drop columns with assessor and OPM remarks (irrelevant to project)

df = df.drop(columns=['Assessor Remarks', 'OPM remarks'])
```

```python
df.to_csv('cleaned_real_estate_data.csv', index=False)
```
